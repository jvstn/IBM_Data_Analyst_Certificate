```python
import pandas as pd
import matplotlib.pylab as plt

filename = "https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/IBMDeveloperSkillsNetwork-DA0101EN-SkillsNetwork/labs/Data%20files/auto.csv"

headers = ["symboling","normalized-losses","make","fuel-type","aspiration", "num-of-doors","body-style",
         "drive-wheels","engine-location","wheel-base", "length","width","height","curb-weight","engine-type",
         "num-of-cylinders", "engine-size","fuel-system","bore","stroke","compression-ratio","horsepower",
         "peak-rpm","city-mpg","highway-mpg","price"]


df = pd.read_csv(filename, names=headers)
df.head(10)
```

Now we want to replace any missing data with represented with "?" 

```python
import numpy as np

df.replace("?", np.nan, inplace=True)
df.head()

```

## Evaluting missing data

Now we can use .isnull() or notnull() method to see all empty values across the dataset

```python
missing_values_df = df.isnull()
missing_values_df.head()
```

```python
for column in missing_values_df.columns.values.tolist():
  # print(column)
  print(missing_values_df[column].value_counts())
  print("")
```

Based on the summary above, each column has 205 rows of data and seven of the columns containing missing data:

"normalized-losses": 41 missing data
"num-of-doors": 2 missing data
"bore": 4 missing data
"stroke" : 4 missing data
"horsepower": 2 missing data
"peak-rpm": 2 missing data
"price": 4 missing data

## Dealing with the missing data

To deal with the missing data we can either replace it with an average for similar entries or we can drop the colum or row.

You only want to drop the column when most of the datapoints for that column are missing

```python
avg_normalized_losses = df["normalized-losses"].astype(float).mean(axis=0)
print(avg_normalized_losses)
df["normalized-losses"].replace(np.nan, avg_normalized_losses, inplace=True)
df["normalized-losses"].head(20)
```

## Replace the rest of missing data

Now we can loop through the rest of our missing columns with missing data and replace with the mean

```python
columns_with_missing_data = ["bore","stroke", "horsepower", "peak-rpm"]

for column in columns_with_missing_data:
  avg_of_column = df[column].astype(float).mean(axis=0)
  print(f"{column} average: {avg_of_column}")
  df[column].replace(np.nan, avg_of_column, inplace=True)
  print(df[column].isnull().value_counts())
```
